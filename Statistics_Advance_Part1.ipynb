**1.  What is a random variable in probability theory?**
- A random variable is a function that assigns numerical values to the outcomes of a random experiment.It helps quantify and analyze uncertain events in probability theory. 

**2. What are the types of random variables?** 
- There are two types:
- **Discrete random variable** Takes countable values like number of heads in coin tosses
- **Continuous random variable** Takes an infinite number of values within a range like height or temperature.

**3. What is the difference between discrete and continuous distributions?** 
- **Discrete distribution:**
- Used for counting integers. e.g. rolling a die.
- Used for distinct events e.g. election, counts.
- Probability functions is Probability Mass function(PMF).
- **Examples of Discrete Distributions:**
- **Bernoulli** Binary outcomes (0 or 1).
- **Binomial** Number of successes in n trials.
- **Poisson** Counts of rare events(e.g. emails/hour).
- **Continuous distribution:**
- Used for measurements (real numbers). e.g. height of a person.
- Used for measurable quantities. e.g. time, distance.
- Probability representation  is Probability Density Function.
- **Examples of Continuous Distributions:**
- **Normal(Gaussian):** Symmetric bell curve.
- **Exponential** Time between events.
- **Uniform** Equal probability over a,b. 

**4. What are probability distribution functions (PDF)?** 
- It is used for continuous random variables.It describes the relative likelihood of a random variable taking on a particular value within a range.
- eg For a normal distribution, the bell curve is the PDF. The probability that a value falls within a certain range  (e.g., 1Ïƒ) is calculated using the area under that part of the curve. 

**5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?** 
- **Probability Distribution Function:**
- It shows density at each point(used for continuous random variables).
- It describes the density of a variable at a specific point.
- f(x) >= 0 and the total area under the curve is 1.
- **Cumulative Distribution Function:**
- It shows the cumulative probability up to a point (for both discrete and continuous). 
- It describes the probability that the variable takes a value less than or equal to x.
- The CDF is always non-decreasing and ranges from 0 to 1.

**6. What is a discrete uniform distribution?** 
- It is a probability distribution where each possible outcome has an equal probability of occuring.It describrs a finite set of values with no bias toward any particular value.

**7. What are the key properties of a Bernoulli distribution?**

**13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?** 
- Regardless of the shape of the original population distribution, the sampling distribution of the mean of a large enough sample will approximate a normal distribution.This holds true as long as the sample size is suffciently large(n>=30) and the data has finite variance.
- **Foundation of Inferential Statistics:** CLT allows us to make predictions and perform hypothesis testing on population parameters using sample data.
- **Justifies the use of normal distribution:** It explains why the normal distribution is used so frequently in statistics even if the population is skewed or unknown.
- **Supports confidence intervals and Z-tests:** It provides the mathematical basis for confidence intervals,Z-test and other statistical tools. 

**14. How does the Central Limit Theorem relate to the normal distribution?** 
- The central limit theorem explains why and when the normal distribution appears in statistics, even if the original data is not normally distributed.
- **1. Sampling Distribution of the Mean:**
- When you take many random samples from any population and calculate the mean of each sample.The distribution of those means will approximate a normal distribution,as long as the sample size is large enough(n>=30)
- **2. Normal Distribution as a Result:**
- Sample means tend to be normally distributed under CLT.This allows us to use z-scores, confidence intervals and hypothesis tests that rely on normality.
- It justifies the use of the normal distribution in:
- Statistical inference
- Confidence intervals
- Hypothesis testing
- Example: Suppose if you're studying customer wait times at a busy store, which are not normally distributed.
- If you take many samples of 30+ customers, calculate the average wait time for each sample, and plot those averages.
- The result will be a bell-shaped (normal) curve because of CLT

**17. What are point estimates and interval estimates in statistics?** 
- **Point estimate:** It is a single value used to estimate an unknown population parameter.
- It is used for quick, specific guesses.
- Example: If the average height of a sample of 100 people is 5.8 feet, we use 5.8 as a point estimate of the population mean height.
- **Interval estimate:** It gives a range of values within which the population parameter is expected to lie, along with a confidence level.
- It is used when we want to include uncertainty and show statistical confidence.
- Example: Instead of saying average height is 5.8 feet, we might say it lies between 5.6 and 6.0 feet with 95% confidence. 

**18. What is the significance of confidence intervals in statistical analysis?** 
- A confidence interval provides a range of values that likely contain the true population parameter (mean or proportion), based on a sample.
- Confidence intervals are crucial in statistics because they quantify the uncertainty in our estimates, helping us make more informed and reliable conclusions.
- **Significance of Confidence interval in statistical analysis**
- **1. Shows Uncertainty Clearly:** Instead of jsut giving one number(point estimate), CI shows the range of possible values, giving a more realistic picture.
- **2. Reflects Realiability:** A 95% confidence interval means that if we repeat the sampling many times, about 95% of the intervals will contain the true population value.
- **3. Helps in Decision-Making:** Confidence intervals guide business decisions, scientific research and policy making by showing how precise or uncertain results are.
- **4. Compares Groups:** If the confidence intervals of two groups do not overlap,we may infer a significant difference between them.
- Example: If we estimate the average salary of data analysts to be 6.5L with a 95% CI of 6.0L and 7.0L  

**19. What is the relationship between a Z-score and a confidence interval?** 
- A Z-score helps determine how far a value is from the mean in terms of standard deviations.In the context of confidence inervals,Z-scores are used to calculate the margin of error.
- The Z-score is used to determine how wide the confidence interval should be, depending on how confident we want to be about our estimate
- Formula to calculate a confidence interval = Point Estimate(Sample Mean) +- Z(Z-score depends on confidence level) * Standard error(Variation in the sample)
- Example: If the sample mean = 50, standard error = 2, and confidence level = 95%
- CI = 50 +- 1.96 * 2 = 50 +- 3.92
- Confidence Interval = (46.08, 53.92)

**20. How are Z-scores used to compare different distributions?**

**21. What are the assumptions for applying the Central Limit Theorem?**
- To apply the Central Limit Theorem correctly, we assume random sampling, independence, a large enough sample size, and that the population has a finite variance
- **1. Random Sampling:** The data should be collected through random and independent sampling.This ensures each data point doesn't influence the others.
- **Sample Size is Large Enough:** The sample size should be sufficiently large(usually n>=30).Larger samples make the sampling distribution closer to normal.
- **Independent Observations:** Each observation in the sample should be independent of others.
- **Finite Variance** The population must have a finite standard deviation and mean.Outliers should not dominate the distribution.
- If the population itself is normal,then the CLT applies even with small sample sizes(n < 30).

**22. What is the concept of expected value in a probability distribution?** 
- Expected value is the weighted average of all possible outcomes, used to predict long-term results of random processes.
- It tells us what outcome you'd expect on average if you repeated an experiment many times.
- Formula: E(X) = âˆ‘[x.P(x)] where x = each possibe outcomes and P(x) = probability of that outcome.
- Example: Tossing a fair 6-sided die
- E(X) = 1+2+3+4+5+6/6 = 21/6 = 3.5
- the expected value is 3.5. The average result over many rolls.

**23. How does a probability distribution relate to the expected outcome of a random variable?** 
- The probability distribution describes all possible value a random variable can take and how likely each one is.
- The expected outcome is the average value you'd get over many trials,based on that distribution.
- The probability distribution tells you the chances of different outcomes,
and the expected value combines those chances to tell you the average result in the long run.
- Example: you have a random variable X representing tossing a biased coin:
- Heads = 1 (with probability 0.7), Tails = 0 (with probability 0.3) (This is the probability distribution.)
- expected value E(X) = 1*0.7 + 0 * 0.3 = 0.7

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
import matplotlib.pyplot as plt 

def bernoulli_pdf(x,p): 
    if x not in [0,1]: 
        raise ValueError("X must be 0 or 1 in a Bernoulli distribution")
    if not (0 <= p <=1): 
        raise ValueError("P must be between 0 and 1")
    return p**x * (1-p)**(1-x)

p = 0.7 
# Probability of success 

x_vals = [0,1]
y_vals = [bernoulli_pdf(x,p) for x in x_vals]

plt.figure(figsize=(6,4))
plt.bar(x_vals, y_vals, color= ["salmon","skyblue"], tick_label=["0(Failure)", "1(Success)"])
plt.title("Outcome")
plt.ylabel("Probability")
plt.ylim(0,1)
plt.grid(axis='y')
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np 
import matplotlib.pyplot as plt 

def calc_and_plot_zscores(data):
    mean = np.mean(data) 
    std = np.std(data) 

    z_scores = (data - mean) / std 

    plt.figure(figsize=(10,5))
    plt.scatter(range(len(z_scores)), z_scores, color='purple', alpha=0.6) 
    plt.axhline(y=0, color='black', linestyle='--', label='Mean (Z=0)')
    plt.axhline(y=1, color='green', linestyle='--', label='Z=1')
    plt.axhline(y=-1, color='red', linestyle='--', label='Z=-1')
    plt.title('Z-scores of the Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

data = np.random.normal(loc=50, scale=10, size = 100)
z_scores = calc_and_plot_zscores(data)
```

```python
#19. Implement hypothesis testing using Z-statistics for a sample dataset. 
# #Suppose a factory claims that the average battery life of their product is 70 hours.
# We randomly sample 50 batteries and find:

# Sample mean = 68.5 hours

# Sample standard deviation = 5.0 hours

# We want to test if the actual mean is different from the claimed mean using a Z-test.
import numpy as np 
from scipy.stats import norm 

claimed_mean = 70 
sample_mean = 68.5
sample_std = 5.0 
n = 50 
alpha = 0.05 

z = (sample_mean - claimed_mean)/(sample_std / np.sqrt(n))
print(f"Z-statistic: {z:.2f}")

p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"P-value: {p_value:.4f}") 

if p_value < alpha: 
    print('Reject the null hypothesis: The avg battery life is significantly  different from 70.')
else: 
    print("Fail to reject the null hypothesis: No significant difference from the claimed 70 hours.")
    
```

```python
#20. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np 
from scipy  import stats 

data = np.array([68, 70, 72, 65, 75, 69, 71, 67, 73, 70])

n = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

confidence_level = 0.95
alpha = 1 - confidence_level

t_critical = stats.t.ppf(1- alpha/2, df=n-1)

margin_error = t_critical * (sample_std / np.sqrt(n))
lower_bound = sample_mean - margin_error
upper_bound = sample_mean + margin_error

print(f"sample mean: {sample_mean: .2f}")
print(f"Sample Standard Deviation: {sample_std: .2f}")
print(f"95% Confidence Interval: ({lower_bound: .2f}, {upper_bound: .2f})")

print("We are 95% confident  that the true mean battery life of the population lies")
print(f"between {lower_bound: .2f} hours and {upper_bound: .2f} hours.")
```

```python
#21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean. 
import numpy as np 
from scipy import stats 

np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=50)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1) 
n = len(data)

confidence = 0.95
alpha = 1- confidence

t_critical = stats.t.ppf(1- alpha/2, df=n-1)
margin_error = t_critical * (sample_std / np.sqrt(n))

lower_bound = sample_mean - margin_error
upper_bound = sample_mean + margin_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"95% Confidence Interval for the Mean: ({lower_bound:.2f}, {upper_bound:.2f})")
print(f"We are 95% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

```python
#25. Write a Python program to compare the standard deviations of two datasets and visualize the difference 
import numpy as np 
import matplotlib.pyplot as plt 

np.random.seed(42)

data1 = np.random.normal(loc=50, scale=5, size=1000)
data2 = np.random.normal(loc=50, scale= 15, size=1000)

std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1: .2f}")
print(f"Standard Deviation of Dataset 2: {std2: .2f}")

plt.figure(figsize=(10,5))
plt.hist(data1, bins=40, alpha=0.6, label='Dataset 1 (Low Std Dev)', color='skyblue', edgecolor='black')
plt.hist(data2, bins=40, alpha=0.6, label='Dataset 2 (High Std Dev)', color='orange', edgecolor='black') 

plt.axvline(np.mean(data1), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(data2), color='darkorange', linestyle='dashed', linewidth=2)

plt.xticks(np.arange(0,101,10))
plt.yticks(np.arange(0,90,5))

plt.title('Comparison of Standard Deviations')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#26 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np 
import matplotlib.pyplot as plt 

data = np.random.normal(loc=50, scale=10, size=1000)

data_range = np.max(data) - np.min(data)
print(f"Range: {data_range: .2f}")

Q1 = np.percentile(data,25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_whisker = np.min(data[data >= Q1 - 1.5 * IQR])
upper_whisker = np.max(data[data <= Q3 + 1.5 * IQR])
print(f"IQR (Interquartile Range): {IQR: .2f}")

plt.boxplot(data, vert=False)
plt.title("Boxplot of the Dataset")
plt.xlabel("Value")

plt.scatter(Q1, 1, color='red', label=f'Q1 = {Q1:.2f}')
plt.scatter(Q3, 1, color='green', label=f'Q3 = {Q3:.2f}')
plt.scatter(lower_whisker, 1, color='blue', label=f'Lower Whisker = {lower_whisker:.2f}')
plt.scatter(upper_whisker, 1, color='purple', label=f'Upper Whisker = {upper_whisker: .2f}')

plt.grid(True)
plt.legend()
plt.show()
```

```python
#27. Implement Z-score normalization on a dataset and visualize its transformation. 
import numpy as np 
import matplotlib.pyplot as plt 

np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=1000)

mean= np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std 

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(data, bins=30, color='skyblue', edgecolor = 'black')
plt.title("Original Data")
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
plt.hist(z_scores, bins=30, color='lightgreen', edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
